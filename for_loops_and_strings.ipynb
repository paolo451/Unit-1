## 1. Multiplication Table

Multiplicationtable.png

```python
x = int(input("Input the number (integer) you wish to know the multiplication table of: "))

msg = f"Multiplication Table for Number {x}"

print(f"\n" + msg)
print("-" * len(msg))

for i in range(1, 11, 1):
  print(f"{x} x {i} = " + str(x*i))
```

## 2. Compound Interest

compoundinterest.png

```python
P = int(input("Initial amount: "))
r = float(input("Interest rate: "))
n = int(input("Number of periods in a year (ex. monthly = 12"))
t = int(input("Period (years)"))

for year in range(t):
  A = P * ((1 + (r/100)/n )) ** (n*year)
  print(f"year {year} amount {A:.2f} US ")
```

## 3. Text Banner

cartel.png

```python
name = str(input("Input the name for the text banner: "))
name_len = len(name)
sides = (48 - name_len) // 2

print("*" * 50)
print("*" + " " * 48 + "*")
if sides % 2 == 0:
  print("*" + " " * sides + name.upper() + " " * sides + "*")
else:
    print("*" + " " * sides + name.upper() + " " * (sides + 1) + "*")
print("*" + " " * 48 + "*")
print("*" * 50)
```
